```
import numpy as np
import pandas as pd
url = "https://raw.githubusercontent.com/vaishnavipatil29/Career-Guidance-ML-Project/master/career_pred.csv"
df1 = pd.read_csv(url)
# Dataset is now stored in a Pandas Dataframe
```

```
df1.head()
```

```
#Data Preprocessing
#1. Handling Missing values:
df1.isna()
```

```
df1=df1.dropna()
df1.isna().sum().sum()
```

```
df1.duplicated()
df1.drop_duplicates()
```

```
df1.dtypes
```

```
df1['can work long time before system?']
```

```
df1['can work long time before system?']=df1['can work long time before system?'].astype('category')
df1['can work long time before system?']=df1['can work long time before system?'].cat.codes
df1['can work long time before system?']
```

```
for col in df1.select_dtypes(include=['object']).columns:
    df1[col] = df1[col].astype('category')
    df1[col] = df1[col].cat.codes
df1.dtypes
```
